```python
from entity_detector import filter_marker_mask, guess_grid_bruteforce, guess_grid_ga
import cv2

img = cv2.imread('tst3.png')
assert img is not None
msk = filter_marker_mask(img, min_brightness=40, min_greenness=30)
g = guess_grid_bruteforce(msk, cell_widths=(32,))
assert g is not None
out = img.copy()

h, w, _ = out.shape
d = 16
for x in g.vl:
    cv2.line(out, (x-d, 0), (x-d, h), (0, 128, 0), 1)
for y in g.hl:
    cv2.line(out, (0, y-d), (w, y-d), (0, 128, 0), 1)
    # for y in g.hl:
        
cv2.imwrite('tst3_.png', out)
cv2.imwrite('msk.png', msk)
print(g)
```
