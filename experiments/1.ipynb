```python
1.0 / 0.032
def time_range(dur: float):
    t0 = time.time()
    i = 0
    grid = [t0]
    while True:
        if len(grid) > 20:
            grid.pop(0)
        t = time.time()
        grid.append(t)
        if len(grid) > 1:
            diffs = [b-a for b, a in list(zip(grid[1:], grid[:-1]))]
            avg_time = sum(diffs) / len(diffs)
        i += 1
        if time.time() - t0 > dur:
            break
        fps = 0.0 if len(grid) < 2 or avg_time == 0 else 1 / avg_time
        yield t, fps, i
```

```python
import psutil

def get_all_processes_command_lines():
    for process in psutil.process_iter(['pid', 'cmdline']):
        try:
            cmdline = process.info['cmdline']
            if cmdline:
                if 'dxgi_text_app' in ' '.join(cmdline):
                    print(f"PID: {process.info['pid']} - Command Line: {' '.join(cmdline)}")
        except (psutil.NoSuchProcess, psutil.AccessDenied, psutil.ZombieProcess):
            pass

if __name__ == "__main__":
    get_all_processes_command_lines()
```
